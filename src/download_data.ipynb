```python
# Open the .npz file in the data folder
import numpy as np
data = np.load('../data/adrenalmnist3d.npz')
```

```python
# Check the keys in the .npz file
print(data.files)
```

```python
# Show some images from the dataset and their labels
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(data['train_images'][i][13], cmap='gray', interpolation='none')
    plt.title("Class {}".format(data['train_labels'][i]))
    
plt.show()
```
